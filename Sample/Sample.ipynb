```python
from pymodelica import compile_fmu
from pyfmi import load_fmu
import matplotlib.pyplot as plt
```

```python
fmu = compile_fmu('SampleAndHold', 'SampleAndHold.mo')
sp = load_fmu(fmu)
res = sp.simulate(final_time=5)
```

```python
t = res['time']
omegaM = res['omegaMeasured']
omega = res['omega1']

plt.figure()
plt.plot(t, omega, label='Omega')
plt.plot(t, omegaM, label='Omega Measured')
plt.xlabel('time(s)')
plt.show()
```

```python
fmu2 = compile_fmu('IntervalMeasure', 'SampleAndHold.mo')
itv = load_fmu(fmu2)
res = itv.simulate(final_time=5)
```

```python
t = res['time']
omega1M = res['omega1Measured']
omega1 = res['omega1']

plt.figure()
plt.plot(t, omega1, label='Omega1')
plt.plot(t, omega1M, label='Omega1 Measured')
plt.xlabel('time(s)')
plt.show()
```

```python
fmu3 = compile_fmu('Counter', 'SampleAndHold.mo')
ct = load_fmu(fmu3)
res3 = ct.simulate(final_time=5)
```

```python
t = res3['time']
omega1M = res3['omega1Measured']
omega1 = res3['omega1']
count = res3['count']

plt.figure(1)
plt.plot(t, omega1, label='Omega1')
plt.plot(t, omega1M, label='Omega1 Measured')
plt.xlabel('time(s)')
plt.figure(2)
plt.plot(t, count)
plt.xlabel('time(s)')
plt.show()
```
